```python
import gensim.downloader as api
from gensim.models import KeyedVectors,Word2Vec
```

```python
# we can import other pretrained models also
# model_name = 'word2vec-google-news-300'
# model = api.load(model_name)
```

```python
# training own model
import logging

# Optional: Setup logging to see the training progress
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
dataset = api.load("text8")
model = Word2Vec(sentences=dataset, vector_size=10, window=3, min_count=5, workers=4)

# Save the model to disk so you don't have to retrain it later
model.save("text8_word2vec.model")
```

```python
# loading saved model
# loaded_model = Word2Vec.load("text8_word2vec.model")
```

```python
# using trainded model
model.wv.most_similar('cricket')
```

```python
model.wv.similarity('man','woman')
```

```python
# odd one out
model.wv.doesnt_match(['python','java','monkey'])
```

```python
vec = model.wv['king'] - model.wv['man'] + model.wv['woman']
model.wv.most_similar([vec])
```
