```python
import string,time
string.punctuation
```

```python
exclude = string.punctuation
def remove_punc(text):
    for char in exclude:
        text = text.replace(char,'') #using replace function
    return text
```

```python
text = 'string. With. Punctuation?'
start = time.time()
print(remove_punc(text))
time1 = time.time() - start   #calculating time
print(time1)
```

```python
def remove_punc1(text):
    return text.translate(str.maketrans('', '', exclude)) # using maketrans

start = time.time()
remove_punc1(text)
time2 = time.time() - start
print(time2)
```

```python
# comparing time
time1/time2  # maketrans is faster
```
