```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = [
    "I love this movie",
    "The movie was boring",
]
cv = CountVectorizer(
        stop_words=None,       # Change to 'english' to remove stop words
        ngram_range=(2, 2),    # Change to (1, 2) for unigram + bigrams
        max_features=None,     # Change to an integer (e.g., 5) to limit vocabulary
        binary=False           # Change to True for binary BoW - max value of each word become 1
    ) 

bow_matrix = cv.fit_transform(corpus)
bow_array = bow_matrix.toarray()
vocab = cv.get_feature_names_out()
    
# Create DataFrame for better visualization
df_sklearn = pd.DataFrame(bow_array, columns=vocab)
print(df_sklearn)
```
