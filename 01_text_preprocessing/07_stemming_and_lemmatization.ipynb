#### Stemming

```python
from nltk.stem.porter import PorterStemmer #for multi lang we use snowball -name is on based on developer

#applied on single word
ps = PorterStemmer()
print(ps.stem("studying"))
print(ps.stem("study"))
```

#### lemmatization

```python
from nltk.stem import WordNetLemmatizer
# nltk.download('wordnet')

ls = WordNetLemmatizer()
print(ls.lemmatize("studying",pos='v')) #need to give what you want to remove
print(ls.lemmatize("study",pos='v'))
```
