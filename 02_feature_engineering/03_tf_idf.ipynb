```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
documents = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "cats and dogs are great"
]
vectorizer = TfidfVectorizer()  #it caclculate log by adding 1 while calculating idf --ensure positive value for common word

# Fit and transform the documents
tfidf_matrix_sklearn = vectorizer.fit_transform(documents)

# Get the feature names (vocabulary)
feature_names = vectorizer.get_feature_names_out()

# Create a DataFrame to display the results nicely (Transposed to match Word x Doc format)
df = pd.DataFrame(
    tfidf_matrix_sklearn.toarray(), 
    index=[f"Doc {i}" for i in range(len(documents))],
    columns=feature_names
)
print(df)
```
